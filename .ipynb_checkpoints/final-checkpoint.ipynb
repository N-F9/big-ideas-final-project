# What Neighborhood is the Best for Pizza Shops?

This is the question we wanted answers to. So, we used various datasets to analytically figure this out. We focused on what a pizza shop would consider the most in picking a location for a pizza shop, the target population, tax friendliness, and the number of pizza shops already in the area. We separately took these 3 main metrics and using the power of statistics made an all-encompassing, weighted statistic. We added the most weight to the target population, then the number of pizza shops, and finally tax friendliness. Pizza shops want to know if they will have a healthy consumer base with little competition, tax friendliness is an added bonus. We also could have considered average annual income or population change, however, we decided to analyse the three above.

```python
# essential imports for the notebook
import pandas as pd
import numpy as np
import geopandas
%matplotlib inline
import matplotlib.pyplot as plt
import math

tax_abatement = pd.read_csv("tax_abatement.csv")
tax_delinquency = pd.read_csv("tax_delinquency.csv")
closed_pizza = pd.read_csv("Sorted_DataSheet.csv")
open_pizza = pd.read_csv("SortedBook2.csv")
pop = pd.read_csv("neighborhoods.csv", index_col= "Neighborhood_2010_HOOD")

neighborhoods_map = geopandas.read_file("Neighborhoods/Neighborhoods_.shp")
```

---

## Submetric 1: Target Population

The target population of a pizza shop is going to be people between the ages of 20 and 60. The data of the kids is irrelevant since it's their parents buying pizza for their family. To find this, we will look at census data of the different neighborhoods of Pittsburgh.

```python
pop['Sum'] = 0.0
pop['tot'] = 0.0
targetpop= pop[["Pop__2010","F__Pop__Age_20_34__2010_","F__Pop__Age_35_59__2010_","Sum",'tot']]

# I created two extra colums. The column 'Sum' is where I will add the percentages of people between 20 and 34, and the percentage of people between 35 and 59, getting the percentage of the population who is between 20 and 60. And the column 'tot' or total, is where I will eventually multiply the sum of percentages by the total population, giving me the true value of people between 20 and 60 in each neighborhood.

for index, row in targetpop.iterrows():
    sum_values = pd.to_numeric(row["F__Pop__Age_20_34__2010_"].rstrip('%'), errors='coerce') + pd.to_numeric(row["F__Pop__Age_35_59__2010_"].rstrip('%'), errors='coerce')
    targetpop.at[index, 'Sum'] = sum_values/100 #creates percentage value

for index, row in targetpop.iterrows():
    tot_values = pd.to_numeric(row['Pop__2010'], errors='coerce') * row["Sum"]
    targetpop.at[index, 'tot'] = math.floor(tot_values)

targetpop=targetpop.sort_values(by='tot',ascending=False)
targetpop.head()
```

Shadyside has the most people between ages 20 and 60 by almost 1000. Therefore, it is the best location for the most people between ages 20 and 60 for pizza shops.

```python
popmap= neighborhoods_map.merge(targetpop, how='left', left_on='hood', right_on='Neighborhood_2010_HOOD')
popmap.plot(column='tot', # set the data to be used for coloring
               cmap='OrRd',              # choose a color palette
               edgecolor="black",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "# of People Age 20-59"}, # label the legend
               figsize=(15, 10),         # set the size
               #missing_kwds={"color": "lightgrey"} # set disctricts with no data to gray
               )
```

From this map, we can see the three best neighborhoods to open up a pizza shop in terms of target population would be **Shadyside**, **Squirrel Hill South**, and **Brookline**.

## Submetric 2: Tax Friendliness

What Neighborhood is the most tax friendly, or in other words, what neighborhood has the least delinquency and the most tax abatements. You want your pizza shop in a strong healthy neighborhood where companies are not falling behind on their taxes and the government actively supporting the local area through tax breaks, this metric is measuring the health of the neighborhood through taxes.

### Criteria for the best neighborhood

- low delinquency amount
- low delinquency prior years
- high abatements amount/year
- the tax metric (encompasses the top three metrics)

---

### Abatements


```python
# make it abatement amount per year averaged across the neighborhoods
tax_abatement["abatement/year"] = tax_abatement['abatement_amount'] / tax_abatement['num_years']
mean_abatement_amount_year = tax_abatement.groupby("neighborhood").mean()['abatement/year']
abatement_amount_map = neighborhoods_map.merge(mean_abatement_amount_year, how='left', left_on='hood', right_on='neighborhood')
abatement_amount_map[['hood','abatement/year']].sort_values('abatement/year', ascending=False).head()
```

Homewood North recieves the most amount per year on average for each abatement, essentially, each abatement they recieve, they will get $33,333.33 on the average. This means that they get a decent amount of tax breaks from the government.

```python
abatement_amount_map.plot(column='abatement/year', # set the data to be used for coloring
               cmap='tab20b',              # choose a color palette
               edgecolor="black",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "average abatement/year"}, # label the legend
               figsize=(15, 10),         # set the size
               missing_kwds={"color": "lightgrey"} # set disctricts with no data to gray
               )
```

---

### Delinquency Amount

```python
mean_delinquency_amount = tax_delinquency.groupby("neighborhood").mean()['current_delq']
delinquency_amount_map = neighborhoods_map.merge(mean_delinquency_amount, how='left', left_on='hood', right_on='neighborhood')
delinquency_amount_map[['hood','current_delq']].sort_values('current_delq').head()
```

Arlington Heights has the lowest average current delinquency of any neighborhood. They are very good about paying taxes on time according to the dataset.

```python
delinquency_amount_map.plot(column='current_delq', # set the data to be used for coloring
               cmap='tab20b',              # choose a color palette
               edgecolor="black",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "average current delinquency"}, # label the legend
               figsize=(15, 10),         # set the size
               missing_kwds={"color": "lightgrey"} # set disctricts with no data to gray
               )
```

---

### Delinquency Prior Years

```python
mean_delinquency_year = tax_delinquency.groupby("neighborhood").mean()['prior_years']
delinquency_year_map = neighborhoods_map.merge(mean_delinquency_year, how='left', left_on='hood', right_on='neighborhood')
delinquency_year_map[['hood','prior_years']].sort_values('prior_years').head()
```

Central Oakland has the lowest prior year delinquency, they pay their delinquencies back the quickest of any neighborhood.

```python
delinquency_year_map.plot(column='prior_years', # set the data to be used for coloring
               cmap='tab20b',              # choose a color palette
               edgecolor="black",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "average prior years"}, # label the legend
               figsize=(15, 10),         # set the size
               missing_kwds={"color": "lightgrey"} # set disctricts with no data to gray
               )
```

---

### Tax Metric

With the tax metric, a higher number is better. This is essentially an all encompassing metric for easy comparison. 

$Tax\space Metric = 0.2(1-(percentile\space ranking\space of\space delinquency\space prior\space years)) + 0.35(percentile\space ranking\space of\space abatements) + 0.45(1-(percentile\space ranking\space of\space delinquency\space amount))$

```python
tax_metric = pd.DataFrame()

tax_metric['delinquency_year_rank'] = 1-mean_delinquency_year.rank(pct=True)
tax_metric['delinquency_amount_rank'] = 1-mean_delinquency_amount.rank(pct=True)
tax_metric['abatement_rank'] = mean_abatement_amount_year.rank(pct=True)
tax_metric = tax_metric.fillna(0)

tax_metric['tax metric'] = .2 * (tax_metric['delinquency_year_rank']) + .35 * (tax_metric['abatement_rank']) + .45 * (tax_metric['delinquency_amount_rank'])

tax_metric.sort_values('tax metric', ascending=False).head()
```

According to the tax metric, Homewood North is the best neighborhood for a pizza shop. It has a good abatement and delinquency year rank, however, it ranks very low in terms of delinquency amount. So, they recieve a good amount of tax cuts and have a hard time with repaying taxes.

```python
tax_metric_map = neighborhoods_map.merge(tax_metric, how='left', left_on='hood', right_on='neighborhood')
tax_metric_map.plot(column='tax metric', # set the data to be used for coloring
               cmap='tab20b',              # choose a color palette
               edgecolor="black",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "tax metric"}, # label the legend
               figsize=(15, 10),         # set the size
               missing_kwds={"color": "lightgrey"} # set disctricts with no data to gray
               )
```

## Submetric 3: Pizza Shops

This metric has datasets of all restruants in Pittsburgh. This data set also shows the activity status of the restrautants. This metric will show where this business has historically not done well.

The best neighborhood is based off of two things:
   * The ratio of closed to open restaurants
   * The number of active restaurants already in the area
 
We will be looking for the neighborhood with the least amount of pizza places and lowest ratio of opened/closed pizza places.

---
### Closed Pizza Places

For this section the data will be represented on a scale from 1 to 7, as that is how the data set marks them as open or closed. We will find the average of the status of all shops in the neighborhoods. The neighborhood with the numbers closest to 7 will have the most closed shops and close to 1 means very little closed shops.

```python
hoods_ratio = closed_pizza.groupby("municipal").mean()['status']
hoods_ratio_map = neighborhoods_map.merge(hoods_ratio, how='left', left_on='hood', right_on='municipal')
hoods_ratio_map[['hood','status']].sort_values('status', ascending=True).head(7)
```

From this table we are able to tell that the neighborhood with the most pizza shop success is East Liberty.

```python
hoods_ratio_map.plot(column='status',
               cmap='YlGn',
               edgecolor="white",
               legend=True,
               legend_kwds={'label': "Likelihood of Closed Pizza Shop"},
               figsize=(15, 10),
               missing_kwds={"color": "black"}
               )
```

This map is also very helpful in out search. It shows that shops in the downtown area have a much higher turnover rate. This is likely due to high rent and running costs in that area. The map also shows it get lighter in the areas around downtown, possibly because these areas still get a lot of traffic from downtown, but nor the same operation costs. Finally, it seems to get dark again on the outskirts of city limits. So we also might want to look for an area closer to downtown, but not in it.

---
### Pizza Competitors
In this section we will try to find the neighborhood with the least amount of competition. More pizza shops means more demand, but less market share.

```python
list_of_un_mun = open_pizza['municipal'].unique()

mun_dict = dict()

for mun in list_of_un_mun:
    count = open_pizza['municipal'].value_counts()[mun]
    mun_dict[mun] = count

keys = mun_dict.keys()
values = mun_dict.values()

data = dict()

data['hood'] = keys
data['open'] = values

mun_df = pd.DataFrame.from_dict(data)
    
hoods_open = open_pizza.groupby("municipal").count()
hoods_open_map = neighborhoods_map.merge(mun_df, how='left', left_on='hood', right_on='hood')
hoods_open_map[['hood','open']].sort_values('open', ascending=False).head()
```

This tells us the worst places to start a business. The worst palce is the North Shore since there are 13 Pizza shops open.

```python
hoods_open_map.plot(column='open',
               cmap='YlGn',
               edgecolor="white",
               legend=True,
               legend_kwds={'label': "Number of Open Pizza Shops"},
               figsize=(15, 10),
               missing_kwds={"color": "black"}
               )
```

We would want to target areas that are lighter in color, if we were going by just this metric. However, we must compare it to our other metrics!

The second set of data tells us where the most active restaurants are. This will help us avoid being suffocated by areas with a lot of pizza.

## Metric: All-Encompassing Weighted

When comparing various metrics, the easiest way to figure out the best neighborhood is through a shared metric to allow for easy comparison. This is what the job of the all encompassing weighted pizza metric (or ANWPM) is for. The formulas for how ANWPM is computed is below. If there is no data for a category, it is simply ignored.

$Population\space Metric = percentile\space ranking\space of\space neighborhood\space population\space between\space 20\space and\space 60$

$Tax\space Metric = 0.2(1-(percentile\space ranking\space of\space delinquency\space prior\space years)) + 0.35(percentile\space ranking\space of\space abatements) + 0.45(1-(percentile\space ranking\space of\space delinquency\space amount))$ 

$Shops\space Metric = .5(1 - percentile\space ranking\space of\space closed\space pizza\space places) + .5(percentile\space ranking\space of\space competitors)$

$ANWPM = .45(Population\space Metric) + .25(Tax\space Metric) + .3(Shops\space Metric)$

```python
population_metric = pd.DataFrame()
mean_population_metric = targetpop.groupby("Neighborhood_2010_HOOD").mean()['tot']
population_metric['population_20_to_60'] = mean_population_metric.rank(pct=True)

shops_metric = pd.DataFrame()
hoods_ratio = closed_pizza.groupby("municipal").mean()['status']
shops_metric['ratio'] = 1-hoods_ratio.rank(pct=True)

open_metric = mun_df.groupby("hood").mean()['open']
shops_metric['shops_open'] = open_metric.rank(pct=True)

shops_metric['ratio'] = shops_metric['ratio'].fillna(0)
shops_metric['shops_open'] = shops_metric['shops_open'].fillna(0)

ANWPM = pd.DataFrame()
ANWPM['population_metric'] = population_metric['population_20_to_60']
ANWPM['tax_metric'] = tax_metric['tax metric']
ANWPM['shops_metric'] = .5 * shops_metric['ratio'] + .5 * shops_metric['shops_open']

ANWPM['population_metric'] = ANWPM['population_metric'].fillna(0)

ANWPM['metric'] = .45 * ANWPM['population_metric'] + .25 * ANWPM['tax_metric'] + .3 * ANWPM['shops_metric'] 
ANWPM.sort_values('metric', ascending=False).head()
```

```python
metric_map = neighborhoods_map.merge(ANWPM, how='left', left_on='hood', right_on='Neighborhood_2010_HOOD')
metric_map.plot(column='metric', # set the data to be used for coloring
               cmap='tab20b',              # choose a color palette
               edgecolor="black",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "ANWPM metric"}, # label the legend
               figsize=(15, 10),         # set the size
               missing_kwds={"color": "lightgrey"} # set disctricts with no data to gray
               )
```
